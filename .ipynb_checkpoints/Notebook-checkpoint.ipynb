```python
import pandas as pd
import numpy as np
import json
import nltk
from nltk.tokenize import word_tokenize
from datetime import datetime

nltk.download('stopwords')
```

```python
#!/usr/bin/env python
# coding: latin-1 

def getMessages(bagOfWorlds, users, timestamps, tokens, date, url):
    f = open(url,)
    data = json.load(f)
    stopword_es = nltk.corpus.stopwords.words('spanish')
    for message in data['messages']:
        if 'content' not in message:
            pass
        else:
            message_txt = message['content']
            message_txt = message_txt.encode('latin1').decode('utf-8').lower()
            

            tokenized_txt = word_tokenize(message_txt)

            for word in tokenized_txt:
                if word in stopword_es:
                    tokenized_txt.remove(word)
                    
            tokens.append(tokenized_txt)
#             tokens.append("")
            
            users.append(message['sender_name'])
            date_val = datetime.fromtimestamp(int(message['timestamp_ms']) / 1000)
            timestamps.append(date_val)
            date.append(datetime.strptime(date_val.strftime("%Y-%m-%d"), "%Y-%m-%d"))
            
            bagOfWorlds.append(message_txt)
#         break

    
```

```python
messages = []
users = []
timestamps = []
tokens = []
date = []
files = 44
# files = 2
for i in range(1, files):
    getMessages(messages, users, timestamps, tokens,date, './facebook-alanhurtarte/messages/inbox/eunicerodas_ajclxnq-w/message_%d.json' % i)

list_of_tuples = list(zip(messages, tokens, users, timestamps, date))
   
df = pd.DataFrame(list_of_tuples,
                  columns = ['message', 'tokens', 'users', 'timestamps', 'date'])
```

```python
import matplotlib.pyplot as plt
# Define a function to plot word cloud
def plot_cloud(wordcloud):
    # Set figure size
    plt.figure(figsize=(40, 30))
    # Display image
    plt.imshow(wordcloud) 
    # No axis details
    plt.axis("off");
```

```python
from sklearn.feature_extraction.text import CountVectorizer
stopword_es = nltk.corpus.stopwords.words('spanish')
count_vect_eu = CountVectorizer(stop_words = stopword_es)
X_eu_train_counts = count_vect_eu.fit_transform(eu_messages['message'])
X_eu_train_counts.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect_alan = CountVectorizer(stop_words = stopword_es)
X_alan_train_counts = count_vect_alan.fit_transform(alan_messages['message'])
X_alan_train_counts.shape
```

```python
def graficarTop(top_words, label):
    labels = []
    values = []
    for word in top_words:
        labels.append(word[0])
        values.append(word[1])

    x = np.arange(len(labels))
    plt.rcParams["figure.figsize"]=15,10
    # plot data in grouped manner of bar type
    plt.bar(x, values, 0.3, color='cyan')
    plt.xticks(x, labels)
    plt.xlabel("Words")
    plt.ylabel("Scores")
    plt.legend([label])
    plt.xticks(fontsize=15, rotation = 45)
    plt.show()
```

```python
sum_words = X_alan_train_counts.sum(axis=0) 
words_freq = [(word, sum_words[0, idx]) for word, idx in     count_vect_alan.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)

words_alan = dict()
for tupla in words_freq:
    words_alan[tupla[0]] = tupla[1]

top_alan = words_freq[:20]
graficarTop(words_freq[:10], 'Alan')
```

```python
sum_words = X_eu_train_counts.sum(axis=0) 
words_freq = [(word, sum_words[0, idx]) for word, idx in     count_vect_eu.vocabulary_.items()]
words_freq =sorted(words_freq, key = lambda x: x[1], reverse=True)

words_eu = dict()
for tupla in words_freq:
    words_eu[tupla[0]] = tupla[1]


top_eu = words_freq[:20]
graficarTop(words_freq[:10], 'Eu')
```

```python
import matplotlib.pyplot as plt
def graficar(top_words1, top_words2, label1, label2):
    labels = []
    values = []
    values2 = []
    for word in top_words1:
        if word[0] not in labels:
            labels.append(word[0])

    for word in top_words2:
        if word[0] not in labels:
            labels.append(word[0])

    for label in labels:
        if label in words_alan:
            values.append(words_alan[label])
        else:
            values.append(0)
            
        if label in words_eu:
            values2.append(words_eu[label])
        else:
            values2.append(0)

    x = np.arange(len(labels))
    plt.rcParams["figure.figsize"]=15,10
    # plot data in grouped manner of bar type
    plt.bar(x-0.3, values, 0.3, color='cyan')
    plt.bar(x, values2, 0.3, color='salmon')
    plt.xticks(x, labels)
    plt.xlabel("Words")
    plt.ylabel("Scores")
    plt.legend([label1, label2])
    plt.xticks(fontsize=15, rotation = 45)
    plt.show()
```

```python
graficar(top_alan, top_eu, 'Alan', 'Eu')
```
